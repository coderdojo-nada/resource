### 円周を積分して円の面積を求める

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

r = 0 # 半径
h = 1 # 半径の増分
integral = 0 # 円周の積分
area = 0 # 円の面積
for i in range(50):
    r += h
    # 円周の長さを面積としてとらえると、円周×線の太さ になる
    # もし、hを0.1にするなら、円周*0.1にする必要がある
    # 足していく円周の太さと、半径の増加分を合わせる必要がある
    # 半径 × 2 × π × 半径の増加分
    integral += r * 2 * np.pi * h  # 円周(面積として扱う)の積分
    area = r**2 * np.pi # その時々の半径での円の面積
    th = np.arange(0, 360)    # 角度 0 - 360
    y = np.sin(np.radians(th))*r
    x = np.cos(np.radians(th))*r
    plt.plot(x, y, "b", lw=0.1)

plt.axis('equal') 
plt.grid(color='0.8')        
plt.show()
print("半径:           " + str(r))
print("円周の積分:" + str(integral))
print("円の面積    :" + str(area))
```

###  円錐の底面積を積分して円錐の体積を求める

```python
%matplotlib nbagg
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
r = 0
h = 1e-4     # 1.0から少数点を4回左にずらす
integral = 0
area = 0
for i in range(0, 10000000):
    r += h
    # 半径 * 半径 * π * 円周の太さ
    integral += r**2 * np.pi * h
    # 半径 * 半径 * 半径 * π  / 3
    area = r**3 * np.pi / 3
    
print("半径:" + str(r))
print("円の面積の積分:" + str(integral))
print("円錐の体積     　　:" + str(area))
```
